## Show the number of confirmed cases , deaths and recovered cases in each region.

```python
covid_df.groupby('Region').sum()
```

## Remove all records where the confirmed cases is less than 10. 

```python
covid_df[~(covid_df.Confirmed<10)]
```

## In which region maximum no of confirmed cases were recorded?

```python
covid_df.groupby('Region').Confirmed.sum().sort_values(ascending = False)
```

## In which region minimum no of deaths were recorded?

```python
covid_df.groupby('Region').Deaths.sum().sort_values(ascending=True)
```

## How many confirmed cases , deaths and recovered cases were reported from India till 29 April 2020?

```python
covid_df[ covid_df.Region == 'India']
```

## Sort the entire data frame with respect to number of recovered cases in descending order 

```python
covid_df.sort_values(by =['Recovered'],ascending=False)
```

## Plot the relationship between Confirmed cases and Deaths using Line graph.

```python
plt.title("Confirmed Cases V/S Deaths")
plt.xlabel("No of Confirmed Cases")
plt.ylabel("No of Deaths")
plt.plot(covid_df.Confirmed,covid_df.Deaths)
plt.figure(figsize=(20,20))
```
